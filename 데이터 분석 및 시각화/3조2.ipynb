반도체 박막 증착 공정은 **웨이퍼 표면에 원자나 분자 수준의 물질을 덧입혀 극도로 얇은 막을 형성하는 기술**로, 이는 반도체를 만드는 8대 핵심 공정 중 하나로서 그 중요성이 강조됩니다.

이 과정의 핵심 목적은 **반도체 칩이 제대로 작동하도록 전기를 차단하는 절연막이나 전기적 경로를 연결하는 전도성막 등 특정한 기능을 수행하는 층을 정밀하게 구축하는 것**입니다.

이러한 정밀도를 확보하기 위해 **PVD, CVD, ALD**와 같은 세 가지 주요 증착 방식이 핵심적으로 활용됩니다.

# ① 화학 증기 증착법(Chemical Vapor Deposition, CVD)

<a href="https://ibb.co/twxnSbqN"><img src="https://i.ibb.co/cKCRPN6p/Gemini-Generated-Image-v8rl6hv8rl6hv8rl.png" alt="Gemini-Generated-Image-v8rl6hv8rl6hv8rl" border="0"></a>

***화학 기상 증착법***

**전구체(Precursor) 가스의 화학 반응을 이용해 웨이퍼 표면에 얇은 막(박막)을 형성하는 방식**입니다.

**공정 과정**

1. 챔버 안으로 전구체 가스가 들어옵니다.
2. 가스들이 반응하며 웨이퍼 표면에 막을 형성합니다.

<br>

***플라즈마 강화 CVD (PECVD)***

**'플라즈마' 상태를 만들어 반응 에너지를 공급**합니다.

**주요 장점**

1. **저온 공정 가능**

    반응에 필요한 온도를 낮출 수 있어, 열에 민감한 금속 배선이나 구조물 위에도 박막을 입힐 수 있습니다.

2. **속도 향상**

    플라즈마의 높은 에너지 덕분에 증착 시간이 줄어듭니다.

3. **고품질**

    더 치밀하고 우수한 품질의 박막을 얻을 수 있습니다.

# ② 원자층 증착법(Atomic Layer Deposition, ALD)

<a href="https://ibb.co/cXk2Pb92"><img src="https://i.ibb.co/WpGBQDXB/Gemini-Generated-Image-8enuwj8enuwj8enu.png" alt="Gemini-Generated-Image-8enuwj8enuwj8enu" border="0"></a>

**원자(Atomic) 두께의 층(Layer)을 한 땀 한 땀 쌓아 올려 박막을 만드는 방식**입니다.

<br>

**핵심 원리**

1. 순차적 단계 (Sequential Steps)

    가스를 한꺼번에 섞어서 반응시키는 CVD와 달리, 단계별로 나누어 진행합니다.

2. 반복 (Cycle)

    '전구체 투입 → (퍼지) → 반응 가스 투입 → (퍼지)'의 과정을 한 세트로 묶어, 이 세트를 반복함으로써 원하는 두께만큼 층을 쌓습니다.

<br>

**주요 특징**

1. 초정밀 두께 제어

    원자 한 층씩 쌓기 때문에 매우 얇고 균일한 막을 만들 수 있습니다.

2. 우수한 피복성 (Step Coverage)

    복잡하고 깊은 구조(단차)에도 구석구석 균일하게 코팅이 가능합니다.

<br>

**단점**

한 층씩 쌓아야 해서 성장 속도가 CVD에 비해 느립니다.

**주목받는 이유**

반도체 소자가 점점 작고 미세해짐(고집적화)에 따라, 나노미터(nm) 단위의 얇고 정교한 막이 필요해졌기 때문입니다.

# ③ 물리적 기상 증착법(Physical Vapor Deposition, PVD)

<a href="https://ibb.co/ks2SKmnq"><img src="https://i.ibb.co/ZpzMd2sc/Gemini-Generated-Image-rnf7crrnf7crrnf7.png" alt="Gemini-Generated-Image-rnf7crrnf7crrnf7" border="0"></a>

***물리 기상 증착 (PVD, Physical Vapor Deposition)***


"화학 반응 없이 물리적인 힘으로 재료를 뜯어내어 덮는 기술"

증착하고자 하는 물질(타겟)을 기체 상태로 만든 뒤, 웨이퍼에 물리적으로 붙이는 방식입니다. 화학 반응을 일으키는 전구체를 쓰지 않는다는 점이 CVD와의 가장 큰 차이입니다.

<br>

***스퍼터링 (Sputtering)***


"이온으로 때려서 튀어나온 입자를 입히는 방식"

핵심 원리
아르곤(Ar) 가스를 플라즈마 상태로 만들어, 이온(Ar+)이 증착 물질(Target)을 물리적으로 타격하게 만듭니다. 이때 충격으로 튀어나온 타겟 입자들이 웨이퍼에 내려앉아 막이 형성됩니다.

<br>

**공정 과정**
1. 플라즈마 생성
    챔버 내에 아르곤(Ar) 가스를 주입하고 전압을 걸어 플라즈마(Ar+)를 만듭니다.

2. 이온 충돌 (Sputtering)
    무거운 아르곤 이온이 강한 힘으로 타겟(증착하려는 금속 덩어리)에 충돌합니다.

3. 박막 형성
    충돌 충격으로 타겟 표면에서 원자가 떨어져 나와(Sputter out), 맞은편에 있는 웨이퍼에 쌓입니다.

<br>

**주요 특징**

1. 금속 증착에 유리
    주로 알루미늄, 구리 같은 금속 배선 공정에 많이 쓰입니다.

2. 고품질
    불순물 오염이 적고 막의 두께 조절이 용이합니다.

3. 직진성
    입자가 직선으로 날아가기 때문에, 복잡한 구조의 구석구석(단차)을 채우는 능력은 CVD나 ALD보다 떨어질 수 있습니다.

# ④ 증착이 잘 됐는지 확인하는 기준(지표)

**1. 균일도 (Uniformity)**

개념: 웨이퍼의 중앙(Center)과 가장자리(Edge) 두께가 얼마나 일정한가?

데이터 관점: 산포도나 표준편차 데이터로 주로 분석합니다.

<br>

**2. 스트레스 (Stress)**

개념: 박막이 웨이퍼를 당기거나(Tensile) 미는(Compressive) 힘.

중요성: 스트레스가 심하면 웨이퍼가 휘거나(Warpage), 막이 갈라질 수 있습니다.

<br>

**3. 굴절률 (Refractive Index, RI)**

개념: 빛이 막을 통과할 때 꺾이는 정도.

중요성: 막의 성분비나 밀도가 정상인지 판단하는 가장 중요한 간접 지표입니다.

(예: 막이 덜 구워졌거나 불순물이 섞이면 굴절률 수치가 튑니다.)

# 3.데이터의 각 공정별 온도 압력 rfpower 가스유량의 평균 또는 중앙값을 조회해서 왜 이런 값을 갖는지 공정과 연결해서 설명해 주세요

```python
import pandas as pd
import numpy as np
import scipy.stats as stats
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.tree import DecisionTreeClassifier, plot_tree, DecisionTreeRegressor
from sklearn.ensemble import RandomForestClassifier, RandomForestRegressor
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
from sklearn.preprocessing import LabelEncoder

# 데이터 로드 및 기본 정보 확인
# CSV 파일 불러오기
df = pd.read_csv('semiconductor.csv')

# 결측치 제거
df_clean = df.dropna().copy()

print("--- 데이터 기본 정보 ---")
print(df.info())
print("\n--- 데이터 상위 5개 행 ---")
print(df.head())

# 분석에 사용할 주요 공정 변수(Feature) 리스트 정의
features = ['process_temp', 'process_pressure', 'process_rf_power', 'gas_flow_rate']
```

```python
#공정별 파라미터(온도, 압력, RF, 유량) 분석
#공정 타입별로 그룹화하여 평균(mean)과 중앙값(median) 계산
process_params = df.groupby('process_type')[features].agg(['mean', 'median'])
print("\n공정별 파라미터 평균 및 중앙값:")
print(process_params)
```

# 4.각 공정별 contamination 결함의 원인을 파악하고 contamination 결함을 방지 하려면 공정을 어떻게 관리해야 하는지 설명해 주세요

```python
# 각 공정별로 반복하며 분석 수행
for proc in ['oxidation', 'etching', 'deposition']:
    # 해당 공정 데이터만 추출
    subset = df_clean[df_clean['process_type'] == proc].copy()
    
    # 목표 설정(y): 결함 타입이 'contamination'이면 1, 아니면 0
    subset['target'] = (subset['defect_type'] == 'contamination').astype(int)
    
    # 오염 데이터가 하나도 없으면 분석 불가하므로 건너뜀
    if subset['target'].sum() == 0:
        print(f"\n[{proc} 공정] 오염 데이터가 없어 분석할 수 없습니다.")
        continue

    # 입력 데이터(X)와 목표 데이터(y) 준비
    X = subset[features]
    y = subset['target']
    
    # Random Forest 모델 학습
    # n_estimators=100: 나무 100그루 심기
    # random_state=42: 매번 똑같은 결과가 나오도록 고정
    rf = RandomForestClassifier(n_estimators=100, random_state=42)
    rf.fit(X, y)
    
    # 변수 중요도 추출 (어떤 변수가 오염에 가장 큰 영향을 미쳤는가?)
    importances = pd.Series(rf.feature_importances_, index=features).sort_values(ascending=False)
    
    # 결과 출력
    print(f"\n[{proc} 공정] 오염 발생 주요 원인 (상위 순서대로):")
    print(importances)
    
    # 가장 중요한 변수 1개를 뽑아서 관리 방안 제안
    top_factor = importances.index[0]
    top_score = importances.iloc[0]
    print(f"결론: {proc} 공정의 오염을 막으려면 '{top_factor}'({top_score*100:.1f}%)를 최우선으로 관리해야 합니다.")
```

# 5.각 공정별 두께에 영향을 미치는 요소를 찾아서 설명해 주세요.

```python
print("--- 공정별 두께(Thickness) 결정 요인 분석 ---")

for proc in ['oxidation', 'etching', 'deposition']:
    subset = df_clean[df_clean['process_type'] == proc]
    X = subset[features]
    y = subset['film_thickness_nm'] # 목표 변수: 두께
    
    # Random Forest 회귀 모델 학습 (두께 예측)
    rf = RandomForestRegressor(n_estimators=100, random_state=42)
    rf.fit(X, y)
    
    # 중요도 확인
    importances = pd.Series(rf.feature_importances_, index=features).sort_values(ascending=False)
    
    # 상관관계 확인 (양의 상관관계 vs 음의 상관관계)
    correlations = subset[features + ['film_thickness_nm']].corr()['film_thickness_nm']
    
    print(f"\n[{proc} 공정] 주요 요인 및 상관계수:")
    print(importances)
    print(f"최고 영향 인자: {importances.index[0]} (상관계수: {correlations[importances.index[0]]:.2f})")
```

# 6.각 공정별 Fab 이나 장비에서 특별히 결함이 많이 생겨서 보수가 필요한Fab이나 장비가 존재 하는지 찾아서 설명해 주세요

```python
# 공정별로 반복 분석
for proc in ['oxidation', 'etching', 'deposition']:
    print(f"\n{'='*20} {proc} 공정 분석 {'='*20}")
    
    # 해당 공정 데이터만 추출
    subset = df[df['process_type'] == proc]
    
    # 장비별 결함 통계 계산 (평균, 합계, 개수)
    # 결함 평균이 높은 순서대로 정렬
    stats_df = subset.groupby(['fab_location', 'equipment'])['defect_count'].agg(['mean', 'std', 'max', 'count'])
    stats_df = stats_df.sort_values(by='mean', ascending=False)
    
    print(f"\n[장비별 결함 현황 Top 5] (평균 결함이 높은 순)")
    print(stats_df.head(5))
    
    # 통계적 검정 (장비 간에 진짜 차이가 있는가?)
    # 장비별로 결함 데이터 리스트 생성
    groups = []
    eq_names = []
    for eq in subset['equipment'].unique():
        data = subset[subset['equipment'] == eq]['defect_count']
        groups.append(data)
        eq_names.append(eq)
        
    # 정규성 검정 (데이터가 적거나 한쪽으로 쏠려있는지 확인)
    is_normal = all([stats.shapiro(g)[1] > 0.05 for g in groups if len(g) >= 3])
    
    # 검정 수행
    if is_normal:
        stat, p = stats.f_oneway(*groups)
        test_name = "ANOVA"
    else:
        stat, p = stats.kruskal(*groups)
        test_name = "Kruskal-Wallis"
        
    print(f"\n>> 통계 검정 결과 ({test_name}): P-value = {p:.5f}")
    if p < 0.05:
        print("   결과: 장비 간에 통계적으로 유의미한 결함 차이가 '있습니다'. (특정 장비가 문제임)")
        problem_eq = stats_df.index[0] # 평균이 가장 높은 장비
        print(f"   보수 필요 장비: {problem_eq} (평균 결함: {stats_df.iloc[0]['mean']:.2f}개)")
    else:
        print("   결과: 장비 간에 결함 차이가 없습니다. (모든 장비가 비슷함)")

    # 시각화 (Boxplot)
    plt.figure(figsize=(10, 6))
    sns.boxplot(data=subset, x='equipment', y='defect_count', hue='fab_location')
    plt.title(f'Defect Count by Equipment ({proc})')
    plt.grid(True, axis='y', alpha=0.5)
    plt.show()
    plt.close()
```

# 7.각 공정의 목표두께는 중앙값으로 설정하고 목표두께에 가깝게 생산하기 위해서 관리되야할 공정 요소가 어떤게 있는지 찾아서 설명해 주세요.

```python
for proc in ['oxidation', 'etching', 'deposition']:
    subset = df_clean[df_clean['process_type'] == proc]
    
    # 목표 두께(중앙값) 설정
    target_thick = subset['film_thickness_nm'].median()
    
    # 두께에 영향을 미치는 변수 중요도 분석 (Random Forest)
    X = subset[features]
    y = subset['film_thickness_nm']
    
    rf = RandomForestRegressor(n_estimators=100, random_state=42)
    rf.fit(X, y)
    
    importances = pd.Series(rf.feature_importances_, index=features).sort_values(ascending=False)
    
    print(f"\n[{proc} 공정]")
    print(f"  - 목표 두께: {target_thick:.2f} nm")
    print(f"  - 최우선 관리 요소: {importances.index[0]} (중요도 {importances.iloc[0]*100:.1f}%)")
    print(f"  - 2순위 관리 요소: {importances.index[1]} (중요도 {importances.iloc[1]*100:.1f}%)")
```
